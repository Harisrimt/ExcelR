```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import warnings
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, r2_score
```

# Q1

```python
df = pd.read_csv("delivery_time.csv")
df.head()
```

```python
df.isna().sum()
```

```python
warnings.filterwarnings("ignore")
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.boxplot(df["Delivery Time"])
plt.subplot(1,2,2)
sns.boxplot(df["Sorting Time"])
```

###### No outliers

```python
sns.pairplot(df)
```

###### The pattern is nearly along a stright line. So, the linear regression can be applied

```python
df.rename(columns = {'Delivery Time':'DT', "Sorting Time": "ST"}, inplace = True)
```

```python
model = smf.ols(formula = 'DT ~ ST' , data = df ).fit()
model.summary()
```

```python
model.params
(model.rsquared,model.rsquared_adj)
```

```python
def calc_rmse(x, y):
    x = sm.add_constant(x)

    model = sm.OLS(y, x).fit()
    
    x_const = sm.add_constant(x)
    y_pred = model.predict(x_const)
    
    mse = mean_squared_error(y, y_pred)  
    rmse = np.sqrt(mse) 
    
    return rmse
```

```python
y = df['DT']
x = df["ST"]
rmse = calc_rmse(x, y)

x_log = np.log(df["ST"])
rmse_log = calc_rmse(x_log, y)

x_sqrt = np.sqrt(df["ST"])
rmse_sqrt = calc_rmse(x_sqrt, y)

x_sq = np.square(df["ST"])
rmse_sq = calc_rmse(x_sq, y)
```

```python
rmse, rmse_log, rmse_sqrt, rmse_sq
```

```python
print("The RMSE value from the prediction is: ", rmse)
print("The RMSE value (after log transformation) from the prediction is: ", rmse_log)
print("The RMSE value (after square root transformation) from the prediction is: ", rmse_sqrt)
print("The RMSE value (after square transformation) from the prediction is: ", rmse_sq)
```

#### Equation is : Delivery time = 1.649 x Sorting time + 6.5827

# Q2

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
df.isna().sum()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.boxplot(df["YearsExperience"])
plt.subplot(1,2,2)
sns.boxplot(df["Salary"])
```

###### No outliers

```python
sns.pairplot(df)
```

###### The relation between input and target variable is fairly linear in nature. Hence linear regression can be applied

```python
model = smf.ols(formula = 'Salary ~ YearsExperience' , data = df ).fit()
model.summary()
```

```python
model.params
(model.rsquared,model.rsquared_adj)
```

#### Equation is : Salary hike = 9449.96 x Years of experience + 25790

```python
def calc_rmse(x, y):
    x = sm.add_constant(x)

    model = sm.OLS(y, x).fit()
    
    x_const = sm.add_constant(x)
    y_pred = model.predict(x_const)
    
    mse = mean_squared_error(y, y_pred)  
    rmse = np.sqrt(mse) 
    
    return rmse
```

```python
y = df['Salary']
x = df["YearsExperience"]
rmse = calc_rmse(x, y)

x_log = np.log(df["YearsExperience"])
rmse_log = calc_rmse(x_log, y)

x_sqrt = np.sqrt(df["YearsExperience"])
rmse_sqrt = calc_rmse(x_sqrt, y)

x_sq = np.square(df["YearsExperience"])
rmse_sq = calc_rmse(x_sq, y)
```

```python
print("The RMSE value from the prediction is: ", rmse)
print("The RMSE value (after log transformation) from the prediction is: ", rmse_log)
print("The RMSE value (after square root transformation) from the prediction is: ", rmse_sqrt)
print("The RMSE value (after square transformation) from the prediction is: ", rmse_sq)
```
