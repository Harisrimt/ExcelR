```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
df_raw = pd.read_csv("book.csv", encoding = 'latin')
df_raw.head()
```

```python
df_raw = df_raw.drop_duplicates()
```

```python
df_raw.shape
```

```python
len(df_raw["Unnamed: 0"].unique()), df_raw.shape
```

```python
df_raw = df_raw.drop(columns = ["Unnamed: 0"], axis = 1)
df_raw.head()
```

```python
df_raw = df_raw.sort_values("User.ID")
df_raw.head()
```

```python
sns.histplot(df_raw["Book.Rating"])
```

Histogram plot shows the frequency of different ratings given to different books. Majority of the ratings are between 7 and 10.

```python
len(df_raw["Book.Title"].value_counts()), len(df_raw["User.ID"].unique())
```

```python
user_books_df = df_raw.pivot_table(index = "User.ID",
                             columns='Book.Title',
                             values='Book.Rating')
user_books_df.head()
```

```python
user_books_df.shape
```

```python
user_books_df.fillna(0, inplace=True)
user_books_df.head()
```

```python
user_books_df.shape
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values, metric='cosine')
```

Cosine similarity shows the similarity of users with the books read and the ratings given to these books. Using the similarity values, we can find out which users have the best correlation or are the most similar in their taste of books so that the books not read by similar users can be suggested according to how the user who has read the book may like it.

###### To directly find cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity( user_books_df.values)
```
